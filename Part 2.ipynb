```python
def recover_digits_from_num(number):
    digit_counter=0
    digits=[]
    temp_num=number
    #print(temp_num)
    while(temp_num>0):
        temp_num=int(temp_num/10)
        digit_counter=digit_counter+1
    #print(digit_counter)
    for x in xrange(1,digit_counter+1):
        if x==digit_counter:
            #print("I've been here")
            digits.append(int(number/10**(x-1)))
        else:
            a=(number-(int(number/10**x))*10**x)
            b=int(a/10**(x-1))
            digits.append(b)
            #print(b)
            #print("Go Go")
    digits=digits[::-1]
    return digits
```

```python
def decimal_to_binary(number):
    binary=[]
    while(number>0):
        if number%2==1:
            binary.append(1)
        elif number==1:
            binary.append(1)
            break
        else:
            binary.append(0)
        number=int(number/2)
    binary=binary[::-1]
    for digit in binary:
        print(digit,end="")
```

```python
def factorial(number):
    factorial=1
    if number==0:
        print("0! equals 1")
    for x in xrange(1,number+1):
        factorial=factorial*x
    print("Factorial of %s is %s"%(number,factorial))
```
